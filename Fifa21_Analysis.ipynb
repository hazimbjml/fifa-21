```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

pd.options.display.max_columns = 100
pd.options.display.max_rows = 100
```

```python
df = pd.read_csv('players_21.csv', sep=',')
```

```python
df.head()
```

```python
#Check missing value
#can also be done by using isnull().any()
df.isnull().sum().sort_values(ascending=False).head(50)
```

```python
df.dtypes
```

```python
df.info(max_cols=110)
```

```python
#Seperating into 2 df, 1 contains only categorical variables (cat_df) and 1 contain only numerical variables (num_df)

cat_df = df.select_dtypes(include=['object']).copy()
cat_df.isnull().sum().sort_values(ascending=False)/cat_df.shape[0]
```

```python
num_df = df.select_dtypes(include=['number']).copy()
num_df.isnull().sum().sort_values(ascending=False)/num_df.shape[0]
```

```python
#each column that has missing values > 30% will be dropped
cat_df.drop(['loaned_from','nation_position','player_tags','player_traits','joined'], axis = 1)
#replace missing with mode for team position, league name, club name
```

```python
num_df.drop(['defending_marking','nation_jersey_number','league_rank','international_reputation','team_jersey_number','gk_kicking','gk_handling','gk_diving','gk_reflexes','gk_speed','gk_positioning'], axis = 1, inplace=True)
#replace missing with mean for physic, dribbling, passing, shooting, pace, defending, release_clause_eur, contract_valid_until
```

```python
num_df = num_df.iloc[ :, 0:18]
num_df
```

```python
#renaming column
num_df.rename(index= int, columns={'sofifa_id':'fifa_id'}, inplace=True)
#remove unnecessary column for this project by  selecting only column index 0-15 for categorical
```

```python
#remove unnecessary column for this project by  selecting only column index 1-14 for categorical

cat_df = cat_df.iloc[ :, 1:14]
cat_df.drop(['body_type','player_tags'], axis = 1, inplace= True)

cat_df
```

```python
num_df
```

```python
#to adjust the size of the plot, it must be written before the plot function
plt.figure(figsize=(30,11))

#mask is used to convernt into triangle corr matrix. np.triu is used to find the matrix
mask = np.triu(num_df.corr())

#vmin/max = set the min & max value of corr. annot is used to display the value
sns.heatmap(num_df.corr(), vmin=-1, vmax=1, annot=True, fmt='.2g', mask=mask, cmap = 'BrBG')

plt.show()
```

```python
plt.figure(figsize=(10,10))
#corr plot for one specific variable
OverallHeatmap = sns.heatmap(num_df.corr()[['overall']].sort_values(by='overall', ascending=False), vmin =-1, vmax=1, annot=True, cmap = 'BrBG')

OverallHeatmap.set_title('Overall Correlation Plot')
```

### The heatmap correlation plot shows that Fifa judged the overall of the player based on the passing and dribbling skill of the player with positive correlation above 0.60.

```python
df = pd.concat([cat_df, num_df], axis = 1)

df
```

```python
SuperLeagueTeams = df[(df['club_name'].isin(['Manchester United','Manchester City','Chelsea',
                                             'Liverpool','FC Barcelona','Real Madrid','Atletico Madrid',
                                             'Inter', 'Juventus','Milan','Paris Saint-Germain',
                                             'Borussia Dortmund','FC Bayern M√ºnchen']))
                         ]
SuperLeagueTeams
```

```python
playerOverall = SuperLeagueTeams[['short_name','overall']]
print(playerOverall.sort_values('overall',ascending=False).head(3))
```

```python
#hue is used to colored the plot based on the categorical variable
#corner is used to make it into triangle so that it wont make any confusion

g = sns.pairplot(SuperLeagueTeams[['overall','wage_eur','passing', 'dribbling','club_name']],hue = 'club_name', corner = True)
```

```python
x = SuperLeagueTeams['wage_eur'].values
result = plt.hist(x, bins=15, color='g')
plt.axvline(x.mean(), color='r', linewidth=2)
plt.show()
print( "\n The average wage is : ", x.mean())
```

```python
max_wage = SuperLeagueTeams.wage_eur.max()
max_wage_player = SuperLeagueTeams[(df['wage_eur'] == max_wage)]['short_name'].values[0]

min_wage = SuperLeagueTeams.wage_eur.min()
min_wage_player = SuperLeagueTeams[(df['wage_eur'] == min_wage)][['short_name','club_name']].values[0]



g = sns.boxplot(y = 'club_name', x = 'wage_eur', data = SuperLeagueTeams, whis=np.inf)

g = sns.swarmplot(y = 'club_name', x = 'wage_eur', data = SuperLeagueTeams, size = 5, color = 'black')

sns.despine() # Used to remove the top and right line in graph

g.figure.set_size_inches(20,10)
plt.xlabel('Wage in Euro (‚Ç¨)')
plt.ylabel('Club Name')
plt.show()

'''plt.annotate(text = max_wage_player, xy = (max_wage, 0), xytext = (500,2), arrowprops = {'facecolor':'red', 'width': 5, 'shrink': 0.8}, backgroundcolor = 'white')
g.figure.set_size_inches(15,9)
plt.show()'''

print("Player with highest wage is:", max_wage_player, max_wage)
print("Player with lowest wage is:", min_wage_player, min_wage)
```
